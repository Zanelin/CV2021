```python
import cv2
import numpy as np
from  matplotlib import pyplot as plt
%matplotlib inline
```

```python
# 讀取圖檔
img = cv2.imread('lena.bmp')
s = img.shape
```

```python
arr = np.zeros(256,dtype=int)
x_axes = np.arange(256)
for i in range(s[0]):
    for j in range(s[1]):
        arr[img[i,j,0]] += 1
plt.subplot(2, 2, 1)
plt.title('origin')
plt.imshow(img.astype('uint8'))
plt.subplot(2, 2, 2)
plt.title('origin historgram')
plt.bar(x_axes,arr) #plt.hist
plt.xlabel('pixel')
plt.show()
```

```python
img2 = img//3
plt.imshow(img2.astype('uint8'))
plt.show()
```

```python
arr2 = np.zeros(256,dtype=int)
for i in range(s[0]):
    for j in range(s[1]):
        arr2[img2[i,j,0]] += 1
plt.bar(x_axes,arr2) #plt.hist
plt.xlabel('pixel')
plt.show()
```

```python
plt.subplot(2, 2, 1)
plt.title('intensity divided by 3')
plt.imshow(img2.astype('uint8'))
plt.subplot(2, 2, 2)
plt.title('intensity divided by 3 historgram')
plt.bar(x_axes,arr2) #plt.hist
plt.xlabel('pixel')
plt.show()
```

```python
histequ = dict()
c = 255/(s[0]*s[1])
for i in range(256):
    histequ[i] = int(sum(arr2[0:i+1])*c)
```

```python
he = np.zeros(img.shape)
for i in range(s[2]):
    for j in range(s[0]):
        for k in range(s[1]):
            he[j,k,i] = histequ[img2[j,k,i]]
plt.imshow(he.astype('uint8'))
plt.show()
```

```python
arr3 = np.zeros(256,dtype=int)
he1 = he.astype(int)
for i in range(s[0]):
    for j in range(s[1]):
        arr3[he1[i,j,0]] += 1
plt.bar(x_axes,arr3) #plt.hist
plt.xlabel('pixel')
plt.show()
```

```python
plt.subplot(2, 2, 1)
plt.title('intensity divided by 3')
plt.imshow(he.astype('uint8'))
plt.subplot(2, 2, 2)
plt.title('intensity divided by 3 historgram')
plt.bar(x_axes,arr3) #plt.hist
plt.xlabel('pixel')
plt.show()
```
