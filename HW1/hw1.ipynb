```python
import cv2
import numpy as np
from  matplotlib import pyplot as plt
%matplotlib inline
```

```python
# 讀取圖檔
img = cv2.imread('lena.bmp')
type(img)
```

```python
s = img.shape
#img.shape[2] #3
print(s)
```

```python
plt.imshow(img)
plt.show()

# 顯示圖片
#cv2.imshow('My Image', img)
# 按下任意鍵則關閉所有視窗
#cv2.waitKey(0)
#cv2.destroyAllWindows()
```

```python
ud = np.zeros(img.shape)
rl = np.zeros(img.shape)
df = np.zeros(img.shape)
print(ud.shape)
```

```python
for i in range(s[2]):
    for j in range(s[0]):
        for k in range(s[1]):
            ud[j,k,i] = img[s[0]-j-1,k,i]
            rl[j,k,i] = img[j,s[1]-k-1,i]
            df[j,k,i] = img[s[0]-j-1,s[1]-k-1,i]
```

```python
plt.imshow(ud.astype('uint8'))
plt.show()
```

```python
plt.imshow(rl.astype('uint8'))
plt.show()
```

```python
plt.imshow(df.astype('uint8'))
plt.show()
```

```python
cv2.imwrite('upside-down_lena.bmp', ud)
cv2.imwrite('right-side-left_lena.bmp', rl)
cv2.imwrite('diagonally_flip_lena.bmp', df)

```

```python
h, w = img.shape[:2]
M = cv2.getRotationMatrix2D((w//2, h//2), -45, 1)
rot = cv2.warpAffine(img, M, (w, h))

plt.imshow(rot.astype('uint8'))
plt.show()
```

```python
shr = np.zeros((img.shape[0]//2,img.shape[1]//2,img.shape[2]))
for i in range(s[2]):
    for j in range(s[0]//2):
        for k in range(s[1]//2):
            shr[j,k,i] = img[j*2,k*2,i]

plt.imshow(shr.astype('uint8'))
plt.show()
```

```python
bn = np.zeros(img.shape)
for i in range(s[2]):
    for j in range(s[0]):
        for k in range(s[1]):
            if img[j,k,i]>128:
                bn[j,k,i] = 255
            else:
                bn[j,k,i] = 0
plt.imshow(bn.astype('uint8'))
plt.show()
```
